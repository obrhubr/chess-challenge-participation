```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
times = [ 
	"Bullet 1+0", #(60000, 0),
	"Bullet 1+1", #(60000, 1000) 
	"Bullet 2+1", #(60000*2, 1000)
	"Blitz 3+0", #(60000*3, 0) 
	"Blitz 3+2", #(60000*3, 1000*2)
	"Blitz 5+0", #(60000*5, 0) 
	"Blitz 5+3", #(60000*5, 1000*3)
];

with open("data.csv") as f:
	i = 0
	for line in f:
		time = list(map(lambda x: int(x), line.split(", ")))
		df = pd.DataFrame(time, index =[x for x in range(0, len(time))],
												columns =['thinkingTime'])
		
		sns.set(rc={'figure.figsize':(11.7,8.27)})
		ax = sns.lineplot(data=df, x=df.index, y="thinkingTime", label=times[i])

		i+=1

	ax.set(xlabel='# of Moves', ylabel='Thinking Time')
	plt.show()
    
```
