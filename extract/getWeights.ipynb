```python
board = chess.Board()
for f in fens:
	board.set_fen(f["fen"])
	# Get King
	positions = board.pieces(6, True)
	for p in positions:
		king[p] = (king[p][0] + f["eval"], king[p][1] + 1)
	positions = board.pieces(6, False)
	for p in positions:
		king[63 - p] = (king[63 - p][0] + f["eval"], king[63 - p][1] + 1)
		
	# Get Queens
	positions = board.pieces(5, True)
	for p in positions:
		queen[p] = (queen[p][0] + f["eval"], queen[p][1] + 1)
	positions = board.pieces(5, False)
	for p in positions:
		queen[63 - p] = (queen[63 - p][0] + f["eval"], queen[63 - p][1] + 1)

	# Get Rooks
	positions = board.pieces(4, True)
	for p in positions:
		rook[p] = (rook[p][0] + f["eval"], rook[p][1] + 1)
	positions = board.pieces(4, False)
	for p in positions:
		rook[63 - p] = (rook[63 - p][0] + f["eval"], rook[63 - p][1] + 1)

	# Get Bishops
	positions = board.pieces(3, True)
	for p in positions:
		bishop[p] = (bishop[p][0] + f["eval"], bishop[p][1] + 1)
	positions = board.pieces(3, False)
	for p in positions:
		bishop[63 - p] = (bishop[63 - p][0] + f["eval"], bishop[63 - p][1] + 1)

	# Get Knights
	positions = board.pieces(2, True)
	for p in positions:
		knight[p] = (knight[p][0] + f["eval"], knight[p][1] + 1)
	positions = board.pieces(2, False)
	for p in positions:
		knight[63 - p] = (knight[63 - p][0] + f["eval"], knight[63 - p][1] + 1)

	# Get Pawns
	positions = board.pieces(1, True)
	for p in positions:
		pawn[p] = (pawn[p][0] + f["eval"], pawn[p][1] + 1)
	positions = board.pieces(1, False)
	for p in positions:
		pawn[63 - p] = (pawn[63 - p][0] + f["eval"], pawn[63 - p][1] + 1)
	break
```

```python
bitboardKing = copy(king)
bitboardQueen = copy(queen)
bitboardRook = copy(rook)
bitboardBishop = copy(bishop)
bitboardKnight = copy(knight)
bitboardPawn = copy(pawn)

for i in range(0, 64):
	if king[i][1] != 0:
		bitboardKing[i] = king[i][0] / king[i][1]
	else:
		bitboardKing[i] = king[i][0]
		
	if queen[i][1] != 0:
		bitboardQueen[i] = queen[i][0] / queen[i][1]
	else:
		bitboardQueen[i] = queen[i][0]

	if rook[i][1] != 0:
		bitboardRook[i] = rook[i][0] / rook[i][1]
	else:
		bitboardRook[i] = rook[i][0]

	if bishop[i][1] != 0:
		bitboardBishop[i] = bishop[i][0] / bishop[i][1]
	else:
		bitboardBishop[i] = bishop[i][0]

	if knight[i][1] != 0:
		bitboardKnight[i] = knight[i][0] / knight[i][1]
	else:
		bitboardKnight[i] = knight[i][0]

	if pawn[i][1] != 0:
		bitboardPawn[i] = pawn[i][0] / pawn[i][1]
	else:
		bitboardPawn[i] = pawn[i][0]
```

```python
jsonString  = json.dumps({
    "Fitness": 0,
    "Weights": {
        "MultiplierKing": bitboardKing,
        "MultiplierQueens": bitboardQueen,
        "MultiplierRooks": bitboardRook,
        "MultiplierBishops": bitboardBishop,
        "MultiplierKnights": bitboardKnight,
        "MultiplierPawns": bitboardPawn,
    }
}, indent=4)

with open("weights.json", "w") as outfile:
    outfile.write(jsonString)
```
