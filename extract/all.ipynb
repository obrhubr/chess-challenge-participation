```python
def setWeights(board, e):
	# Get King
	positions = board.pieces(6, True)
	for p in positions:
		king[p] = (king[p][0] + e, king[p][1] + 1)
	positions = board.pieces(6, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		king[index] = (king[index][0] + e, king[index][1] + 1)
		
	# Get Queens
	positions = board.pieces(5, True)
	for p in positions:
		queen[p] = (queen[p][0] + e, queen[p][1] + 1)
	positions = board.pieces(5, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		queen[index] = (queen[index][0] + e, queen[index][1] + 1)

	# Get Rooks
	positions = board.pieces(4, True)
	for p in positions:
		rook[p] = (rook[p][0] + e, rook[p][1] + 1)
	positions = board.pieces(4, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		rook[index] = (rook[index][0] + e, rook[index][1] + 1)

	# Get Bishops
	positions = board.pieces(3, True)
	for p in positions:
		bishop[p] = (bishop[p][0] + e, bishop[p][1] + 1)
	positions = board.pieces(3, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		bishop[index] = (bishop[index][0] + e, bishop[index][1] + 1)

	# Get Knights
	positions = board.pieces(2, True)
	for p in positions:
		knight[p] = (knight[p][0] + e, knight[p][1] + 1)
	positions = board.pieces(2, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		knight[index] = (knight[index][0] + e, knight[index][1] + 1)

	# Get Pawns
	positions = board.pieces(1, True)
	for p in positions:
		pawn[p] = (pawn[p][0] + e, pawn[p][1] + 1)
	positions = board.pieces(1, False)
	for p in positions:
		index = ((63 - p) // 8 + 1) * 8 - ((63 - p) % 8) - 1
		pawn[index] = (pawn[index][0] + e, pawn[index][1] + 1)
```

```python
def setBitboards():
	for i in range(0, 64):
		if king[i][1] != 0:
			bitboardKing[i] = king[i][0] / king[i][1]
		else:
			bitboardKing[i] = king[i][0]
			
		if queen[i][1] != 0:
			bitboardQueen[i] = queen[i][0] / queen[i][1]
		else:
			bitboardQueen[i] = queen[i][0]

		if rook[i][1] != 0:
			bitboardRook[i] = rook[i][0] / rook[i][1]
		else:
			bitboardRook[i] = rook[i][0]

		if bishop[i][1] != 0:
			bitboardBishop[i] = bishop[i][0] / bishop[i][1]
		else:
			bitboardBishop[i] = bishop[i][0]

		if knight[i][1] != 0:
			bitboardKnight[i] = knight[i][0] / knight[i][1]
		else:
			bitboardKnight[i] = knight[i][0]

		if pawn[i][1] != 0:
			bitboardPawn[i] = pawn[i][0] / pawn[i][1]
		else:
			bitboardPawn[i] = pawn[i][0]
```

```python
def writeToFile():
	jsonStringWeights = json.dumps({
		"Fitness": 0,
		"Weights": {
			"MultiplierKing": bitboardKing,
			"MultiplierQueens": bitboardQueen,
			"MultiplierRooks": bitboardRook,
			"MultiplierBishops": bitboardBishop,
			"MultiplierKnights": bitboardKnight,
			"MultiplierPawns": bitboardPawn,
		}
	}, indent=4)

	kingExport = [0] * 64
	for i in range(0,64): 
		kingExport[i] = { "val": king[i][0], "multi": king[i][1] }
	queenExport = [0] * 64
	for i in range(0,64): 
		queenExport[i] = { "val": queen[i][0], "multi": queen[i][1] }
	rookExport = [0] * 64
	for i in range(0,64): 
		rookExport[i] = { "val": rook[i][0], "multi": rook[i][1] }
	bishopExport = [0] * 64
	for i in range(0,64): 
		bishopExport[i] = { "val": bishop[i][0], "multi": bishop[i][1] }
	knightExport = [0] * 64
	for i in range(0,64): 
		knightExport[i] = { "val": knight[i][0], "multi": knight[i][1] }
	pawnExport = [0] * 64
	for i in range(0,64): 
		pawnExport[i] = { "val": pawn[i][0], "multi": pawn[i][1] }
	
	jsonStringWorking = json.dumps({
		"Weights": {
			"king": kingExport,
			"queen": queenExport,
			"rook": rookExport,
			"bishop": bishopExport,
			"knight": knightExport,
			"pawn": pawnExport,
		}
	}, indent=4)

	with open("weights.json", "w") as outfile:
		outfile.write(jsonStringWeights)
	with open("working.json", "w") as outfile:
		outfile.write(jsonStringWorking)

	# Output to c#
	bitboards = {
		"MultiplierKing": bitboardKing,
		"MultiplierQueens": bitboardQueen,
		"MultiplierRooks": bitboardRook,
		"MultiplierBishops": bitboardBishop,
		"MultiplierKnights": bitboardKnight,
		"MultiplierPawns": bitboardPawn,
	}
	with open("weights.cs", "w") as outfile:
		for key in bitboards:
			outfile.writelines("static double[] " + key + " = {\n")
			line = str(bitboards[key][0:8])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][8:16])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][16:24])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][24:32])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][32:40])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][40:48])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][48:56])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			line = str(bitboards[key][56:64])
			outfile.writelines("    " + line[1:len(line)-1] + ",\n")
			outfile.writelines("};\n")
```

```python
# skip already trained games
with open("lichess_db_standard_rated_2023-07.pgn") as pgn:
	for j in range(20 * king[0][1]):
		if j % 500 == 0:
			print("Skipped lines: " + str(j))
		next(pgn)
	
	print("Skipped " + str(king[0][1]) + " games in total. Beginning training...")

	i = king[0][1]

	while True:
		mygame = chess.pgn.read_game(pgn)

		if mygame is None:
			break
		
		# Check if the game has eval
		board = mygame.next()
		if board is None:
			continue
		p = re.compile(r"\[%eval ((\+|-)?[0-9]*\.[0-9]*)")
		e = p.search(board.comment)

		if e is not None:
			while mygame.next():
				board = mygame.board()
				if board.is_checkmate():
					continue
				
				p = re.compile(r"\[%eval ((\+|-)?[0-9]*\.[0-9]*)")
				e = p.search(mygame.comment)
				if e is not None:
					setWeights(board, float(e.group(1)))
				
				mygame = mygame.next()

			i += 1
			if i % 50 == 0:
				print("Game number: " + str(i))
				setBitboards()
				writeToFile()
```
