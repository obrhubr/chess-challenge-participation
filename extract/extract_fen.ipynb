```python
import chess.pgn
import json
import re
```

```python
pgn = open("filtered.pgn")
fens = []
```

```python
while True:
    mygame = chess.pgn.read_game(pgn)
    if mygame is None:
        break

    stockfishColor = mygame.headers["White"].__contains__("Stockfish")
    print(mygame.headers["White"] + " vs. " + mygame.headers["Black"])
    while mygame.next():
        if mygame.turn() == stockfishColor:
            mygame = mygame.next()
            continue

        board = mygame.board()
        if board.is_checkmate():
            continue
        
        p = re.compile(r"\[%eval ((\+|-)?[0-9]*\.[0-9]*)")

        eval = p.search(mygame.comment)
        if eval is not None:
            eval = eval.group(1)
            fens += [{ "fen": board.fen(), "eval": float(eval) }]
        
        mygame = mygame.next()

json_object = json.dumps({"positions": fens}, indent=4)
 
# Writing to sample.json
with open("analysis.json", "w") as outfile:
    outfile.write(json_object)
```
